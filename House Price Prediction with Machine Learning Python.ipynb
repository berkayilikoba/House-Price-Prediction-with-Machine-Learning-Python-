```python
import pandas as pd 
data = pd.read_csv('dataset.csv')
data
```

```python
data.drop(columns=['Id'], inplace=True)
data
```

```python
object_columns = data.select_dtypes(include=['object'])

for col in object_columns:
    if data[col].isnull().sum() > 0:
        print(col, data[col].isnull().sum())
```

```python
object_columns = data.select_dtypes(include=['object'])

for col in object_columns:
    if data[col].isnull().sum() < data.shape[1]:
        data[col] = data[col].fillna(data[col].mode()[0])
        
    elif data[col].isnull().sum() > data.shape[1]:
        data.drop(columns=[col], inplace=True)
        
data.shape
```

```python
object_cols = data.select_dtypes(include=['object'])
object_cols.nunique()
```

```python
for col in object_cols:
    if data[col].nunique() > 3:
        mean_prices = data.groupby(col)['SalePrice'].mean()
        overall_median = mean_prices.median()
        data['Categorized '+ col] =  data[col].map(lambda brand: 1 if mean_prices[brand] > overall_median else 0)
        data.drop(columns=[col], inplace=True)
    
    elif data[col].nunique() == 2:
        from sklearn.preprocessing import LabelEncoder
            
        le = LabelEncoder()
        data[col] = le.fit_transform(data[col])
    
    elif data[col].nunique() == 3:
        from sklearn.preprocessing import OneHotEncoder
        ohe = OneHotEncoder(drop='first')
        input_data = data[col].values.reshape(-1, 1)  
        encoded_cols = ohe.fit_transform(input_data)
        col_names = [f"{data.columns[0]}_{category}" for category in ohe.categories_[0] if category != ohe.categories_[0][0]]
        encoded_df = pd.DataFrame(encoded_cols.toarray(), columns=col_names)
        data = data.drop(col, axis=1)
        data = pd.concat([data, encoded_df], axis=1)
            
data
```

```python
corrs = data.corr()['SalePrice'].drop('SalePrice')
vals = corrs[abs(corrs) < 0.5].index.tolist()
    
for v in vals:
    data.drop(columns=[v], inplace=True)
    print(v)
    lenght = len(vals)
    print(lenght , "feature was deleted")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

y = data.iloc[:, 10:11].values
y = pd.DataFrame(data=y, columns=['SalePrice'])

cp = data.copy()

cp.drop(columns=['SalePrice'], inplace=True)

x = pd.DataFrame(data=cp)


x_train, x_test, y_train, y_test = train_test_split(x,y)

lin = LinearRegression()
lin.fit(x_train, y_train)

from sklearn.metrics import r2_score

p1 = lin.predict(x_train)
p2 = lin.predict(x_test)

r1 = r2_score(y_train, p1)
r2 = r2_score(y_test, p2)

print(r1)
print(r2)
```
